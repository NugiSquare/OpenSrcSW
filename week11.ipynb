```python
#Assignment9

class Stack:
    def __init__(self):
        self.stack = []

    def stackpush(self, num):
        self.stack.append(num)
    
    def stackpop(self):
        return self.stack.pop()
        
    def stacklength(self):
        return len(self.stack)
    
    def printcontent(self):
        print("The content is: ")
        for val in self.stack:
            print(val)

def calculator(int1, op1, int2):
    if op1 == "+":
        return str(int(int1) + int(int2))
    elif op1 == "-":
        return str(int(int1) - int(int2))
    elif op1 == "*":
        return str(int(int1) * int(int2))
    elif op1 == "/":
        return str(int(int1) / int(int2))
    else:
        print("Wrong operator, Abort.")

strinput = input().split()

num = Stack() #num stack
op = Stack()  #op stack

#As Professor's explanation
for idx, val in enumerate(strinput):
    if (idx % 2 == 0):
        num.stackpush(val)
    else:
        op.stackpush(val)
    if (num.stacklength() == 2):
        int1 = num.stackpop()
        int2 = num.stackpop()
        op1 = op.stackpop()
        num.stackpush(calculator(int1, op1, int2))
        
for idx, val in enumerate(strinput):
    print(val, end='')
print("=" + str(num.stackpop()))
        
#As Right Calculation
```
