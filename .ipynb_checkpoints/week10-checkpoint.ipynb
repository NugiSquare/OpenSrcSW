```python
#Assignment3

print("Input String:")
strinput = input().split()

print("Press 1 for encryption, Press 2 for decryption:")
menuinput = int(input())
resultlist = []
if menuinput == 1:
    for a in strinput:
        tmplist1 = list(a)
        tmplist2 = []
        for b in tmplist1:
            tmplist2.append(chr(ord(b) + 1))
        resultlist.append(''.join(tmplist2))
    print(' '.join(list(reversed(resultlist))))
elif menuinput == 2:
    for a in strinput:
        tmplist1 = list(a)
        tmplist2 = []
        for b in tmplist1:
            tmplist2.append(chr(ord(b) - 1))
        resultlist.append(''.join(tmplist2))
    print(' '.join(list(reversed(resultlist))))
else:
    print("Wrong Input.")
```

```python
#Assignment4

data = {}
print("10 People name and phone number, please.")

i = 1
while i<= 10:
    strarr = input().split()
    data[strarr[0]] = strarr[1]
    i += 1
    
print("Press 1 for searching by name, Press 2 for searching by phone number:")
menuinput = int(input())
if menuinput == 1:
    print(data[input()])
elif menuinput == 2:
    print(list(data.keys())[list(data.values()).index(input())])
else:
    print("Wrong Input.")
```
