```python
from collections import Counter
import csv

f = open('tmoney.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

metro_dict = {}

for row in data:
        metro_dict[row[3]] = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))

k = Counter(metro_dict)
        
# Finding 3 highest values
high = k.most_common(3) 
  
print('이용객이 가장 많은 역 순위')
for i in high:
    print(i[0]," :",i[1]," ")

# Finding 3 lowest values
low = list(reversed(k.most_common()))
print(low)
low = low[:3]

print('이용객이 가장 적은 역 순위')
for i in low:
    print(i[0]," :",i[1]," ")
```
