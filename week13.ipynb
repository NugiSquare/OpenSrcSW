```python
#Assignment 1

import matplotlib.pyplot as plt
import csv

f = open('ta_20210525191158.csv', 'r', encoding='cp949')

data = csv.reader(f)
next(data)
result_highest = []
result_lowest = []
for row in data:
    if row[-1] != '':
        if row[0].split('.')[1] == '3' and row[0].split('.')[2] == '23':
            result_highest.append(float(row[4]))
            result_lowest.append(float(row[3]))

plt.title('Birthday Temperature')

plt.plot(result_highest, color='pink', label='Highest Temperature')
plt.plot(result_lowest, color='skyblue', label='Lowest Temperature')
plt.legend()

plt.show()
```

```python
#Assignment 2

import matplotlib.pyplot as plt
import csv

f = open('ta_20210525191158.csv', 'r', encoding='cp949')

data = csv.reader(f)
next(data)
result_jan = []
result_aug = []
for row in data:
    if row[-1] != '':
        if row[0].split('.')[1] == '1':
            result_jan.append(float(row[4]))
        elif row[0].split('.')[1] == '8':
            result_aug.append(float(row[4]))

plt.title('Highest Temperature during JAN and AUG')

#plt.plot(result_jan, color='blue', label='JAN')
#plt.plot(result_aug, color='red', label='AUG')
plt.hist(result_jan, bins=100, color='b', label='JAN')
plt.hist(result_aug, bins=100, color='r', label='AUG')
plt.legend()

plt.show()
```
